```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

order_items_df = pd.read_csv("order_items_dataset.csv")
order_reviews_df = pd.read_csv("order_reviews_dataset.csv")
products_df = pd.read_csv("products_dataset.csv")
products_translate_df = pd.read_csv("product_category_name_translation.csv")

merged_df = pd.merge(order_items_df, products_df, on="product_id", how="inner")

merged_df = pd.merge(merged_df, products_translate_df, on="product_category_name", how="left")

merged_df["product_category_name"] = merged_df["product_category_name_english"]

category_sales = merged_df.groupby("product_category_name")["order_id"].count().reset_index()

category_sales.columns = ["product_category_name", "total_sales"]

category_sales = category_sales.sort_values(by="total_sales", ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x="total_sales", 
            y="product_category_name", 
            data=category_sales.head(10),  
            hue="product_category_name",  
            palette="viridis",
            legend=False)

plt.xlabel("Jumlah Penjualan")
plt.ylabel("Kategori Produk")
plt.title("10 Kategori Produk dengan Penjualan Terbanyak")
plt.gca().invert_yaxis()
plt.show()

merged_reviews_df = pd.merge(order_reviews_df, merged_df[["order_id", "product_id", "product_category_name"]], on="order_id", how="inner")


product_reviews = merged_reviews_df.groupby(["product_id", "product_category_name"]).agg(
    total_reviews=("review_score", "count")
).reset_index()

most_reviewed_products = product_reviews.sort_values(by="total_reviews", ascending=False)

top_3_products = most_reviewed_products.head(3)["product_id"].tolist()

for product in top_3_products:
    review_counts = merged_reviews_df[merged_reviews_df["product_id"] == product]["review_score"].value_counts()

    most_common_rating = review_counts.idxmax()

plt.figure(figsize=(12, 6))
sns.barplot(
    data=most_reviewed_products.head(10), 
    x="total_reviews", 
    y="product_id",  
    hue="product_category_name",  
    palette="magma", 
    legend=False
)
plt.xlabel("Total Reviews")
plt.ylabel("Product ID")
plt.title("Top 10 Category with Most Reviews")
plt.show()

plt.figure(figsize=(12, 6))

for product in top_3_products:
    review_counts = merged_reviews_df[merged_reviews_df["product_id"] == product]["review_score"].value_counts()
    sns.barplot(x=review_counts.index, y=review_counts.values, label=f"Product ID {product}")

plt.xlabel("Review Score")
plt.ylabel("Count")
plt.title("Review Score Distribution for Top 3 Most Reviewed Products")
plt.legend()
plt.show()

orders_df = pd.read_csv("orders_dataset.csv")
customers_df = pd.read_csv("customers_dataset.csv")

merged_city_df = pd.merge(orders_df, customers_df, on="customer_id", how="inner")

city_high = merged_city_df.groupby("customer_city")["customer_id"].count().reset_index()

city_high.columns = ["customer_city", "total_orders"]

city_high = city_high.sort_values(by="total_orders", ascending=False)

top_10_cities = city_high.head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_cities["total_orders"], 
            y=top_10_cities["customer_city"], 
            hue=top_10_cities["customer_city"],  
            palette="Blues_r", 
            legend=False) 

plt.xlabel("Total Orders")
plt.ylabel("Customer City")
plt.title("Top 10 Cities with the Most Orders")
plt.show()

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

order_items_df = pd.read_csv("order_items_dataset.csv")
orders_df = pd.read_csv("orders_dataset.csv")

orders_df["order_date"] = pd.to_datetime(orders_df["order_purchase_timestamp"])

order_items_df["total_price"] = order_items_df["price"] + order_items_df["freight_value"]

all_df = pd.merge(order_items_df, orders_df, on="order_id", how="inner")

rfm_df = all_df.groupby(by="customer_id", as_index=False).agg({
    "order_date": "max",     
    "order_id": "nunique",    
    "total_price": "sum"      
})

rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]

rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = orders_df["order_date"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x: (recent_date - x).days)

rfm_df.drop("max_order_timestamp", axis=1, inplace=True)

def replace_customer_labels(df, column, top_n=5):
    """Mengganti customer_id dengan label A, B, C, D, E dan mencetak mapping-nya."""
    top_customers = df.sort_values(by=column, ascending=(column == "recency")).head(top_n)
    mapping = {customer: label for customer, label in zip(top_customers["customer_id"], ["A", "B", "C", "D", "E"])}
    df["customer_label"] = df["customer_id"].map(mapping)
    return df, mapping

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(30, 6))

colors = ["#72BCD4"] * 5

rfm_df, recency_mapping = replace_customer_labels(rfm_df, "recency")
sns.barplot(y="recency", x="customer_label", hue="customer_label", data=rfm_df.sort_values(by="recency", ascending=True).head(5), palette=colors, legend=False, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("By Recency (days)", loc="center", fontsize=18)
ax[0].tick_params(axis='x', labelsize=15)
ax[0].set_ylim(top=rfm_df["recency"].max())   


rfm_df, frequency_mapping = replace_customer_labels(rfm_df, "frequency")
sns.barplot(y="frequency", x="customer_label", hue="customer_label", data=rfm_df.sort_values(by="frequency", ascending=False).head(5), palette=colors, legend=False, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].set_title("By Frequency", loc="center", fontsize=18)
ax[1].tick_params(axis='x', labelsize=15)
ax[1].set_ylim(top=rfm_df["frequency"].max() + 1) 

rfm_df, monetary_mapping = replace_customer_labels(rfm_df, "monetary")
sns.barplot(y="monetary", x="customer_label", hue="customer_label", data=rfm_df.sort_values(by="monetary", ascending=False).head(5), palette=colors, legend=False, ax=ax[2])
ax[2].set_ylabel(None)
ax[2].set_xlabel(None)
ax[2].set_title("By Monetary", loc="center", fontsize=18)
ax[2].tick_params(axis='x', labelsize=15)

plt.suptitle("Best Customer Based on RFM Parameters", fontsize=20)
plt.show()

print("Customer ID:")
for key, value in recency_mapping.items():
    print(f"{value}  {key}")



```
