```python
# Import library yang diperlukan
import pandas as pd
import geopandas as gpd
import folium
from folium.plugins import HeatMap
import matplotlib.pyplot as plt
import geodatasets  # Library baru untuk dataset geospasial

# 1. Membaca data CSV
df = pd.read_csv('geolocation_dataset.csv')

# 2. Membuat GeoDataFrame dari latitude dan longitude
gdf = gpd.GeoDataFrame(
    df, 
    geometry=gpd.points_from_xy(df.geolocation_lng, df.geolocation_lat),
    crs="EPSG:4326"
)

# 3. Analisis Dasar
print("Jumlah data per kota:")
print(df['geolocation_city'].value_counts())

# 4. Membuat peta dasar dengan Folium
center_lat = df['geolocation_lat'].mean()
center_lng = df['geolocation_lng'].mean()

m = folium.Map(location=[center_lat, center_lng], 
               zoom_start=10,
               tiles='OpenStreetMap')

# 5. Menambahkan marker untuk setiap titik
for idx, row in gdf.iterrows():
    folium.Marker(
        location=[row['geolocation_lat'], row['geolocation_lng']],
        popup=row['geolocation_city'],
        icon=folium.Icon(color='blue')
    ).add_to(m)

# 6. Membuat Heatmap
heat_data = [[row['geolocation_lat'], row['geolocation_lng']] 
            for idx, row in df.iterrows()]
HeatMap(heat_data).add_to(m)

# 7. Visualisasi tambahan menggunakan GeoPandas
fig, ax = plt.subplots(figsize=(10, 10))
# Menggunakan dataset dari geodatasets
world = gpd.read_file(geodatasets.get_path('natural_earth.land'))
world.plot(ax=ax, color='lightgrey')
gdf.plot(ax=ax, color='red', markersize=5, alpha=0.5)
plt.title('Distribusi Spasial Data')
plt.show()

# 8. Menyimpan peta sebagai HTML
m.save('geospatial_map.html')

print("Analisis selesai! Peta telah disimpan sebagai 'geospatial_map.html'")
```
