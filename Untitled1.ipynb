```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
orders_df = pd.read_csv("orders_dataset.csv")

# Hitung jumlah transaksi per pelanggan (customer_id)
transaction_counts = orders_df.groupby("customer_id")["order_id"].nunique().reset_index()
transaction_counts.columns = ["customer_id", "transaction_count"]

# Terapkan teknik binning untuk clustering transaksi
bins = [0, 1, 3, 6, 10, np.inf]
labels = ["Very Low", "Low", "Medium", "High", "Very High"]
transaction_counts["transaction_cluster"] = pd.cut(transaction_counts["transaction_count"], bins=bins, labels=labels, right=False)

# Cek distribusi hasil clustering
print("Distribusi cluster transaksi:")
print(transaction_counts["transaction_cluster"].value_counts())

# Visualisasi hasil clustering
plt.figure(figsize=(10, 6))
sns.countplot(x="transaction_cluster", data=transaction_counts, palette="coolwarm")
plt.xlabel("Transaction Cluster")
plt.ylabel("Count of Customers")
plt.title("Customer Clustering Based on Transaction Count")
plt.show()

# Tampilkan beberapa contoh data hasil clustering
print(transaction_counts.sample(10))  # Menampilkan 20 data acak untuk pemeriksaan
```
